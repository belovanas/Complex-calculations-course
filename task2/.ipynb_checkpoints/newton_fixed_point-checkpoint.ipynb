# Part I. Root-finding. Newton's iteration.

Write a function which performs Newton's iteration for a given function $f(x)$ with known derivative $f'(x)$. Your function should find the root of $f(x)$ with a predefined absolute accuracy $\epsilon$. 

```python
import warnings
import numpy as np
import matplotlib.pyplot as plt
from math import pi


def newton_iteration(f, fder, x0, eps=1e-5, maxiter=1000):
    """Find a root of $f(x) = 0$ via Newton's iteration starting from x0.
    
    Parameters
    ----------
    f : callable
        The function to find a root of.
    fder : callable
        The derivative of `f`.
    x0 : float
        Initial value for the Newton's iteration.
    eps : float
        The target accuracy. 
        The iteration stops when the distance between successive iterates is below `eps`.
        Default is 1e-5.
    maxiter : int
        The maximum number of iterations (default is 1000.)
        Iterations terminate if the number of iterations exceeds `maxiter`.
        This parameter is only needed to avoid infinite loops if iterations wander off.
        
    Returns
    -------
    x : float
        The estimate for the root.
    niter : int
        The number of iterations.
    """
    niter = 1
    xold = x0
    while niter < maxiter:
        xnew = xold - f(xold)/fder(xold)
        if abs(xnew - xold) < eps: break
        xold = xnew
        niter += 1
        
    if niter == maxiter:
        warnings.warn("Convergence not fulfilled after {:d} iterations".format(maxiter))
    
    return xnew,niter
```

### Test I.1 

Test your implementation on a simple example, $f(x) = x^2 - 1$ or similar. (20% of the total grade)

```python
# define function and derivative
func = lambda x: x**2 - 1.0
fder = lambda x: 2.0*x

# set initial guesses
x1_0 = -14.88273
x2_0 =  73.12742

# find xstar1 = -1.0
x1,n1 = newton_iteration(func, fder, x1_0, eps=1e-9, maxiter=10)
print("Found root x = {: f}, at iteration {:>3d}".format(x1,n1))

# find xstar2 = +1.0
x2,n2 = newton_iteration(func, fder, x2_0, eps=1e-9, maxiter=20)
print("Found root x = {: f}, at iteration {:>3d}".format(x2,n2))
```

### Test I.2

Now consider a function which has a multiple root. Take $f(x) = (x^2 - 1)^2$ as an example. Implement a modified Newton's iteration,

$$
x_{n+1} = x_{n} - m \frac{f(x_n)}{f'(x_n)}
$$

and vary $m= 1, 2, 3, 4, 5$. Check the number of iterations required for convergence within a fixed $\epsilon$. Are your observations consistent with the expectation that the convergence is quadratic if $m$ equals the multiplicity of the root, and is linear otherwise? (40% of the total grade)

```python
def newton_iteration_mult(f, fder, x0, m=1, eps=1e-5, maxiter=1000, series_root=False):
    """Find a root of $f(x) = 0$ via Newton's iteration starting from x0.
    
    Parameters
    ----------
    f : callable
        The function to find a root of.
    fder : callable
        The derivative of `f`.
    x0 : float
        Initial value for the Newton's iteration.
    m : int
        Multiplicity of the root sought (default is 1).
    eps : float
        The target accuracy. 
        The iteration stops when the distance between successive iterates is below `eps`.
        Default is 1e-5.
    maxiter : int
        The maximum number of iterations (default is 1000)
        Iterations terminate if the number of iterations exceeds `maxiter`.
        This parameter is only needed to avoid infinite loops if iterations wander off.
    series_root : boolean
        A flag that allows to return the array containing the whole series of the
        approximations of the root (default is False)
        
    Returns
    -------
    xseries : np.array(float)
        The series of the approximations of the root.
    x : float
        The estimate for the root.
    niter : int
        The number of iterations.
    """
    m = float(m)
    niter = 1
    xold = x0
    
    if series_root:
        xseries = np.empty(shape=(maxiter,))
        xseries.fill(np.nan)
    else:
        xseries = np.empty(shape=(1,))
    
    xseries[0] = xold
    while niter < maxiter:
        xnew = xold - m*f(xold)/fder(xold)
        if series_root: xseries[niter] = xnew
        if abs(xnew - xold) < eps: break
        xold = xnew
        niter += 1
        
    if niter == maxiter:
        warnings.warn("Convergence not fulfilled after {:d} iterations".format(maxiter))
    
    return xseries[~np.isnan(xseries)],xnew,niter
```

```python
# define function and derivative
func = lambda x: (x**2 - 1.0)**2
fder = lambda x: 4.0*(x**2 - 1.0)*x

# set solver parameters
x_0 = -1.77865
maxiter = 80
eps = 1.0e-4

# find xstar1 = -1.0, for different values of m
x0 = np.zeros(shape=(5,), dtype='float64')
x = np.zeros(shape=(5,), dtype='float64')
n = np.zeros(shape=(5,), dtype='uint16')
for m in range(1,6):
    x0[m-1],x[m-1],n[m-1] = newton_iteration_mult(func, fder, x_0, m, eps=eps, maxiter=maxiter)
    print("Found root x = {: f} for m = {:d}, at iteration {:>4d}".format(x[m-1],m,n[m-1]))
    
# plot convergence of |x^(n+1) - x*| / |x^(n) - x*|
xstar = -1.0
x = np.zeros(shape=(5,), dtype='float64')
n = np.zeros(shape=(5,), dtype='uint16')
f,axs = plt.subplots(2,3, figsize=(15, 6))
f.subplots_adjust(hspace = .5)
axs = axs.ravel()
for m in range(1,6):
    xseries,x[m-1],n[m-1] = newton_iteration_mult(func, fder, x_0, m, eps=eps, maxiter=maxiter, series_root=True)
    if m != 2:
        delta = np.abs(xseries[1:] - xstar) / np.abs(xseries[:-1] - xstar)
    else:
        delta = np.abs(xseries[1:] - xstar) / np.abs(xseries[:-1] - xstar)**2
    t = np.arange(np.size(delta)) + 1
    axs[m-1].plot(t,delta)
    axs[m-1].set_title("m = {:d}".format(m))
```

The above plots illustrate the profile of the ratio of the distance of successive iteration of the root of $(x^2 - 1)^2 = 0$ (the analytical zero sought is $x^* = -1$), as defined:

$$\varepsilon_n=\dfrac{|x_{n+1}-x^*|}{|x_n-x^*|^{\alpha}}$$

The value of $\alpha$ adopted is 1, for $m\neq2$, and 2 for $m=2$. Therefore, in the case of $m=2$ the convergence is quadratic since the asymptotic error constant is $\frac{1}{2}$ for $\alpha=2$, whilst the convergence is linear for $m=1,3$ since the asymptotic error constant is $\frac{1}{2}$ for $\alpha=1$. The algorithm does not converge for $m=4,5$.

# Part II. Fixed-point iteration

Consider the following equation:

$$
\sqrt{x} = \cos{x}
$$

Plot the left-hand side and right-hand side of this equation, and localize the root graphically. Estimate the location of the root by visual inspection of the plot.

Write a function which finds the solution using fixed-point iteration up to a predefined accuracy $\epsilon$. Compare the result to an estimate from a visual inspection.

Next, rewrite the fixed-point problem in the form

$$
x = x - \alpha f(x)
$$

where $\alpha$ is the free parameter. Check the dependence of the number of iterations required for a given $\epsilon$ on $\alpha$. Compare your results to an expectation that the optimal value of $\alpha$ is given by 

$$
\alpha = \frac{2}{m + M}
$$

where $0 < m < |f'(x)| < M$ over the localization interval. (40% of the total grade)

```python
# define the plot domain
x = np.linspace(0., 2*pi, 501)
ycosx = np.cos(x)
ysqrt = np.sqrt(x)

# plot the solution
f = plt.figure(0, figsize=(9,6))
plt.title("Plot of $\sqrt{x}$ and $\cos{(x)}$")
plt.plot(x, ysqrt, 'b-', label='$\sqrt{x}$')
plt.plot(x, ycosx, 'r-', label='$\cos{(x)}$')
plt.legend(loc='upper center')
plt.show()
```

```python
yc = lambda x: np.cos(x)
ys = lambda x: np.sqrt(x)

# visual inspection of the solution
x_visual = 0.65
print( "sqrt(0.65) ={: f}".format(yc(x_visual)) )
print( "cos(0.65)  ={: f}".format(ys(x_visual)) )
```

```python
def fixed_point(phi, x0, eps=1e-6, maxiter=1000):
    """Find a root of $f(x) = 0$ via the fixed-point iteration method in
    the form $x = phi(x)$, starting from x0.
    
    Parameters
    ----------
    phi : callable
        The fixed-point function.
    x0 : float
        Initial value for the Newton's iteration.
    eps : float
        The target accuracy. 
        The iteration stops when the distance between successive iterates is below `eps`.
        Default is 1e-6.
    maxiter : int
        The maximum number of iterations (default is 1000)
        Iterations terminate if the number of iterations exceeds `maxiter`.
        This parameter is only needed to avoid infinite loops if iterations wander off.
        
    Returns
    -------
    x : float
        The estimate for the root.
    niter : int
        The number of iterations.
    """
    niter = 1
    xold = x0
    while niter < maxiter:
        xnew = phi(xold)
        if abs(xnew - xold) < eps: break
        xold = xnew
        niter += 1
        
    if niter == maxiter:
        warnings.warn("Convergence not fulfilled after {:d} iterations".format(maxiter))
    
    return xnew,niter
```

```python
# define the fixed-point function
phi = lambda x: x - np.sqrt(x) + np.cos(x)

# solve $\sqrt{x} = \cos{x}$ by means of the fixed-point iteration method defined above
xfp,nfp = fixed_point(phi, 6.0, maxiter=40)

# comparison with the visual inspection
print("Fixed-point solution (eps = 1E-06) x = {: f} vs visual solution x = {: f}".format(xfp,x_visual))
```

```python
def fixed_point_improved(f, x0, alpha=1.0, eps=1e-6, maxiter=1000):
    """Find a root of $f(x) = 0$ via the fixed-point iteration method in
    the form $x = x - alpha*f(x)$, starting from x0.
    
    Parameters
    ----------
    f : callable
        The root-finiding function.
    x0 : float
        Initial value for the Newton's iteration.
    alpha : float
        The multiplier of the function $f(x)$.
    eps : float
        The target accuracy. 
        The iteration stops when the distance between successive iterates is below `eps`.
        Default is 1e-6.
    maxiter : int
        The maximum number of iterations (default is 1000)
        Iterations terminate if the number of iterations exceeds `maxiter`.
        This parameter is only needed to avoid infinite loops if iterations wander off.
        
    Returns
    -------
    x : float
        The estimate for the root.
    niter : int
        The number of iterations.
    """
    niter = 1
    xold = x0
    while niter < maxiter:
        xnew = xold - alpha*f(xold)
        if abs(xnew - xold) < eps: break
        xold = xnew
        niter += 1
        
    if niter == maxiter:
        warnings.warn("Convergence not fulfilled after {:d} iterations".format(maxiter))
    
    return xnew,niter
```

```python
# define the root-finding function and derivative
funct = lambda x: np.sqrt(x) - np.cos(x)
funcd = lambda x: 0.5/np.sqrt(x) + np.sin(x)

# solve $x = x - \alpha*(\cos{x} - \sqrt{x})$ by means of the improved fixed-point iteration
# method defined above
xfpi,nfpi = fixed_point_improved(funct, 6.0, alpha = 1.0)

# define the localization interval [a,b]
a = 0.6
b = 0.7
nint = int( (b-a)*10 + 1 )

# find min & max of f' in [a,b]
xd = np.linspace(a, b, nint)
fi = np.abs( funcd(xd) )
m = np.min(fi)
M = np.max(fi)

# solution for the minimizing value of alpha
amin = 2.0/(m+M)
x_amin,n_amin = fixed_point_improved(funct, 0.7, alpha = amin)

# check dependency of the algorithm w.r.t. alpha
alpha = np.linspace(0.1, 1.4, 14)
adim = len(alpha)
x_a = np.zeros(adim, dtype='float64')
n_a = np.zeros(adim, dtype='int')
for k in range(adim):
    x_a[k],n_a[k] = fixed_point_improved(funct, 0.7, alpha = alpha[k])

# plot dependency of iterations vs alpha
f1 = plt.figure(1, figsize=(9,6))
plt.title(r"Fixed-point iterations vs $\alpha$")
plt.xlabel(r'$\alpha$')
plt.ylabel('Number of iterations')
plt.plot(alpha, n_a, 'b.', label=r'$\alpha = [0.1,2.0]$')
plt.plot(amin, n_amin, 'rx', label=r'$\alpha_{min}$')
plt.show()
```

# Part III. Newton's fractal.

(Not graded). 

Consider the equation

$$
x^3 = 1
$$

It has three solutions in the complex plane, $x_k = \exp(i\, 2\pi k/ 3)$, $k = 0, 1, 2$.

The Newton's iterations converge to one of these solutions, depending on the starting point in the complex plane (to converge to a complex-valued solution, the iteration needs a complex-valued starting point).

Plot the \emph{basins of attraction} of these roots on the complex plane of $x$ (i.e., on the plane $\mathrm{Re}x$ -- $\mathrm{Im}x$). To this end, make a series of calculations, varying the initial conditions on a grid of points. 
Color the grid in three colors, according to the root, to which iterations converged.

```python
# define function and derivative
func = lambda x: x**3 - 1.0
fder = lambda x: 3.0*x**2

# analytical roots
x = np.array([1.0, complex(-0.5,np.sqrt(3)/2.), complex(-0.5,-np.sqrt(3)/2.)])
z = np.array([0, 1, 2], dtype=int)

# set solution space
im = np.linspace(-5., 5., 501)
re = np.linspace(-5., 5., 501)

# compute roots
y = np.empty( shape=(np.size(re),np.size(im)), dtype=int )
for i in range(np.size(re)):
    for j in range(np.size(im)):
        if ((re[i] == 0.) & (im[j] == 0.)): continue
        xij,n = newton_iteration(func, fder, complex(re[i],im[j]), eps=1e-6, maxiter=100)
        norm = ( (x - xij).conj() * (x - xij) ).real
        y[i,j] = z[norm < 1.0e-10]

# plot basins of attraction
f,ax = plt.subplots()
h = ax.contourf(re, im, y, levels=[-0.5,0.5,1.5,2.5])
proxy = [plt.Rectangle((0,0),1,1,fc = pc.get_facecolor()[0]) for pc in h.collections]
plt.legend(proxy, [r"{$x_0: x_n \rightarrow 1$}", r"{$x_0: x_n \rightarrow -0.5 + \mathrm{i}\sqrt{3}/2$}", 
                   r"{$x_0: x_n \rightarrow -0.5 - \mathrm{i}\sqrt{3}/2$}"], loc='center left',
           bbox_to_anchor=(1, 0.5))
ax.set_title("Newton's fractals of the solutions of $x^3 = 1$")
ax.set_xlabel('Re($x$)')
ax.set_ylabel('Im($x$)')

plt.show()
```
